```python
from graphics import *
from random import *
from time import *
```

```python
class Snake:
    def __init__(self, snake_list):
        self.snake_list = snake_list

    def move(self, key):
        'The function is responsible for the movement. By implementing the key parameter which returns the last key pressed '
        'It also saves the snake positions and adds to its list (snake_list)'
        global x, y, a, length_of_snake
        x_change = 0
        y_change = 1
        if a == "w":
            x_change = 0
            y_change = 1
        elif a == "s":
            x_change = 0
            y_change = -1
        elif a == "d":
            x_change = 1
            y_change = 0
        elif a == "a":
            x_change = -1
            y_change = 0
        if key == '':
            a = a
        else:
            a = key
        x += x_change
        y += y_change
        self.snake_list.append([x, y])
        if len(self.snake_list) > length_of_snake:  #Removal of already present snake leaf elements
            del self.snake_list[0]

    def draw(self):
        'The function is responsible for creating snake body based on snake_list'
        for i in range(0, len(self.snake_list)):
            r = Rectangle(Point(self.snake_list[i][0], self.snake_list[i][1]),
                          Point(self.snake_list[i][0] - 1, self.snake_list[i][1] - 1))
            r.setFill('black')
            r.draw(win)

    def undraw(self):
        'This function is used to "redraw" the snake by creating new kefels in old places with a white background.'
        for x in range(0, len(self.snake_list)):
            r = Rectangle(Point(self.snake_list[x][0], self.snake_list[x][1]),
                          Point(self.snake_list[x][0] - 1, self.snake_list[x][1] - 1))
            r.setFill('white')
            r.setOutline('white')
            r.draw(win)

    def lose(self):
        'The function looks for every possibility to end game'
        global game_close
        if x > 20 or x <= 0 or y > 20 or y < 0:  # If player go out the map game_lose = 0
            game_close = 0
        if len(self.snake_list) > 1:  # If player hit snake body game is over
            for v in self.snake_list[:-1]:
                if v == [x, y]:
                    game_close = 0
```

```python
def message():
    'Initial screen to start the game'
    message = Text(Point(10, 10), 'Press any key to start.')
    message.draw(win)
    win.getKey()
    message.undraw()
```

```python
def main():
    'Main function responsible for the operation of the entire game'
    global win, x, y, a, game_close, food_x, food_y, food, length_of_snake

    win = GraphWin('Snake', 500, 500)
    win.setBackground('white')
    win.setCoords(0, 0, 20, 20)

    game_close = 1  # If game_close = 1 the game is still running

    a = ' '         # Variable which holds the last key pressed in memory
    key = win.checkKey()
    x = 10          # x, y variables that tell us about the position of the "snake's head".
    y = 10
    food_x = randint(1, 20)     # food_x, food_y The variables that hold the location of the snake's food in memory
    food_y = randint(1, 20)
    length_of_snake = 1         # The length of the snake that tells you how many tiles the snake consists of
    s = Snake([[10, 10]])
    message()

    food = Circle(Point(food_x - 1 / 2, food_y - 1 / 2), 1 / 3)     # Creating food for the snake
    food.setFill("red")
    food.draw(win)

    while game_close:
        t = Text(Point(1.5, 19.5), f'Score {length_of_snake}')  # creating a score in the upper left corner of the game
        t.draw(win)

        key = win.checkKey()
        s.move(key)
        s.draw()
        sleep(1 / 8)            # Slow down the game
        s.undraw()
        s.lose()

        if x == food_x and y == food_y:
            'In case of hitting the food snake, the following commands are triggered and they' 
            'redraw the previous ones food and draw a new place.'

            food.undraw()
            food_x = randint(1, 20)
            food_y = randint(1, 20)
            food = Circle(Point(food_x - 1 / 2, food_y - 1 / 2), 1 / 3)
            food.setFill("red")
            food.draw(win)
            length_of_snake += 1        # Increasing the length of the snake by 1, another tile is added

        t.undraw()                      # undraw the punctation

        if game_close == 0:
            'If the game is lost, the result will appear'
            food.undraw()
            message2 = Text(Point(10, 10), f'Your score {length_of_snake}')
            message2.setSize(25)
            message2.setTextColor('red')
            message2.draw(win)
            win.getKey()

    win.close()

main()
```
